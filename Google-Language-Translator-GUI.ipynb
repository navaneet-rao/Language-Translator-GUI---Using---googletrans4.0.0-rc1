```python
import googletrans
from googletrans import Translator
from tkinter import *
```

This are all the languages supposed by Google Translater 

```python
googleLangs = googletrans.LANGUAGES
print(googleLangs)
```

```python
try:
    translator = Translator(service_urls=['translate.google.com','translate.google.co.kr',])
except ReadTimeout:
    print("ReadTimeout occurred. Waiting for 5 seconds before retrying...")
    time.sleep(20)  # pause for 8 seconds    
```

```python
options = googleLangs.values()
```

```python
win = Tk()
win.geometry("750x650")
win.title("Language Translator GUI")
```

```python
userTextLabel = Label(win, text="Enter Text:", font=("Verdana", 10))
userTextLabel.pack(pady=10)

userTextBox = Text(font=("Verdana", 10), width=100, height=10)
userTextBox.pack(padx=5, pady=5)
```

```python
def InputText():
#     print(userTextBox.get(1.0, "end-1c"))
    return (userTextBox.get(1.0, "end-1c"))

def selectedLang():
    userSelectedLang = clicked.get()
#     print(userSelectedLang)
    label.config(text=f"Language you have choose is : {userSelectedLang}")
    languageKey = {i for i in googleLangs if googleLangs[i] == userSelectedLang }
#     print("key by value:", languageKey)
    return languageKey
```

Eexception Handling for Unsupported Language and Network Issues  

```python
def translateUserText():
    try:
        source_text = InputText()
        language_set = selectedLang()
        
        language_str = language_set.pop()
        
        TranslatedTextBox.delete("1.0", END)
        
        translated_result = translator.translate(source_text, dest=language_str)
        TranslatedTextBox.insert("1.0", translated_result.text)
        
    except ValueError as ve:
        if 'unsupported language' in str(ve):
            TranslatedTextBox.insert("1.0", "Error: Unsupported language '{source_text}' or '{language_str}'.")
        
    except Exception as e:
        print(f"Translation error: {e}")
        TranslatedTextBox.insert("1.0","Please select a valid Language and Enter your text also check your Network")
```

```python
clicked = StringVar() 
clicked.set( "Languages" ) 
```

```python
menuFrame = Frame(win)

menuFrame.columnconfigure(0, weight=1)
menuFrame.columnconfigure(2, weight=1)

drop_lang = OptionMenu( menuFrame ,clicked , *options)
drop_lang.configure(font=("Verdana", 10))
drop_lang.grid(row=0, column=0, padx=5, pady=5) 
btn_trans = Button(menuFrame, text='Translate', font=("Verdana", 10), command = lambda: [InputText(), selectedLang(), translateUserText()] )
btn_trans.grid(row=0, column=1, padx=5, pady=5)

menuFrame.pack(padx=5)

label = Label( win , text = " " ) 
label.pack() 
```

```python
TranslatedTextLabel = Label(win, text="Translated Text:", font=("Verdana", 10))
TranslatedTextLabel.pack(pady=10)

TranslatedTextBox = Text(win, font=("Verdana", 10), width=100, height=10)
TranslatedTextBox.pack(padx=5, pady=5)
```

```python
win.mainloop()
```
